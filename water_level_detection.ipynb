```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('water2.DNG')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
gray = cv2.resize(gray, (300, 500))
```

```python
ret, thresh = cv2.threshold(gray, 10, 90, cv2.THRESH_BINARY)
```

```python
cv2_imshow(thresh)
```

```python
print(thresh)
```

```python
thresh.mean()
```

```python
thresh.shape
```

```python
clahe = cv2.createCLAHE(clipLimit=1000.0, tileGridSize=(2, 2))
```

```python
enhanced = clahe.apply(gray)
```

```python
# blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
canny = cv2.Canny(enhanced, 50, 150)
```

```python
cv2_imshow(enhanced)
```
